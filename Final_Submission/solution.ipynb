```python
import csv
```

```python
path2="/content/drive/MyDrive/recommendations"
path3="/content/drive/MyDrive/Final_Persons_And_Recommenders.csv"
list2=os.listdir(path2)
with open(path3, 'r') as file:
    reader=csv.reader(file)
    data={}
    for row in reader:
      data[row[0]]=row[1]
suspicious=[]
for i in data:
  recommenders = data[i].strip('[]').split(', ')
  for j in recommenders:
      if j in data and i in data[j].strip('[]').split(', '):
        if i not in suspicious:
          suspicious.append(i)
        if j not in suspicious:
          suspicious.append(j)
print(suspicious)

```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

def rate_vagueness(resume_text):
    if isinstance(resume_text, (str, bytes)) and resume_text:
        tfidf_vectorizer = TfidfVectorizer()
        tfidf_matrix = tfidf_vectorizer.fit_transform([resume_text])
        # Calculate the average TF-IDF score (rest of your function logic)
        avg_tfidf = tfidf_matrix.mean(axis=1)[0]

        # Assign a score based on average TF-IDF
        score=0
        l=list((0.3-0.01*x for x in range(0,200)))
        for i in range(len(l)):
          if avg_tfidf<l[i]:
            score=len(l)-i-1
          else:
            break

        return score*5
    else:
        # Return a default value for invalid inputs like np.nan or empty strings
        return 0

```

```python
dframe['score_5']=(dframe['skills'].apply(lambda x: " ".join(x))).apply(rate_vagueness)
dframe['score_5']=dframe['score_5'].apply(int)/200
```

```python
recomm=pd.read_csv('/content/drive/MyDrive/cleaned.csv')
recomm['recommendations'] = recomm['recommendations'].fillna('')
recomm['score_5']=recomm['recommendations'].apply(rate_vagueness)
recomm['score_5']=recomm['score_5'].apply(lambda x: x-855)
print(recomm['score_5'])
```

```python
dframe['cycles']=0
for i in range(len(dframe)):
  if str(i) in suspicious:
    dframe['cycles'][i]=-1
  else:
    dframe['cycles'][i]=1
```

```python
cv_sc=list(dframe['score_5'])
rec_sc=list(recomm['score_5'])
cycles=list(dframe['cycles'])
final_sc=[]
print(len(cv_sc))
print(len(rec_sc))
print(len(cycles))
for i in range(len(cycles)):
  if i < len(rec_sc):
    final_sc.append(cv_sc[i]*10+rec_sc[i]*cycles[i])
  else:
    final_sc.append(cv_sc[i]*10)
print(final_sc)
print(len(final_sc))
```

```python
dframe['final_score']=0
for j in range(len(final_sc)):
  dframe['final_score'][j]=final_sc[j]
dframe1=dframe.sort_values('final_score',ascending=False)
```
